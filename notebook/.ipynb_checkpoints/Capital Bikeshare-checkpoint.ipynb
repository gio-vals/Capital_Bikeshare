### calculation of 95% confidence interval for the mean of the windspeed feature using bootstraping and z-value formula

```python
# bootstrapping.

df_windspeed = df['windspeed']
df_windspeed
```

```python
# creating an empty list called sample_means

sample_means_list = []
```

```python
# Creating a for loop that loops 10,000 times and in each iteration 100 windspeed samples will be held 
# and their mean is appended in the empty list from above

for i in range(10_000):
    bootstrap_sample = df_windspeed.sample(n=100, replace=True) 
    sample_means_list.append(bootstrap_sample.mean())
```

```python
# plotting the distribution of the sample_means list

sns.displot(data=sample_means_list, bins=40);
plt.xlabel('windspeed');
```

```python
sns.displot(data=sample_means_list, kind='kde');
plt.xlabel('windspeed');
```

```python
# Converting sample_means list to pandas Series

sample_means_series = pd.Series(sample_means_list)
```

```python
# Using pandas .quantile() calculate the upper (0.975) and 
# lower (0.025) limits of the confidence interval of the sample_means Series

clvl = 0.95

lower_quantile = (1-clvl)/2  

upper_quantile = clvl + ((1-clvl)/2)

print(clvl, lower_quantile, upper_quantile)
```

```python
lower_limit = sample_means_series.quantile(lower_quantile)

upper_limit = sample_means_series.quantile(upper_quantile)

print(f'The Bootstrapped Confidence Interval of the average windspeed is between {lower_limit} and {upper_limit}')
```

```python
# Calculate using the z-value formula

# sample mean and std 

# calculate mean of bill length for Adelie:
sample_mean = df_windspeed.mean()

# calculate std of bill length for Adelie:
sample_std = df_windspeed.std()

print(sample_mean, sample_std)
```

```python
# clvl = 0.95, then z = 1.96

z = 1.96
n = len(df_windspeed)

lower_limit_eq = sample_mean - ((z*sample_std)/np.sqrt(n))
upper_limit_eq = sample_mean + ((z*sample_std)/np.sqrt(n))

print(f'The Calculated (z value) Confidence Interval of the average windspeed is between {lower_limit_eq} and {upper_limit_eq}')
```

```python
# the same output with one line of code!

sms.DescrStatsW(df_windspeed).tconfint_mean()
```

```python
# Comparing to the results of the bootstrapping method

# the z method is better
```

```python
plt.figure(figsize=(14,6))

sns.histplot(sample_means_list, kde=True, bins=40)

plt.axvline(lower_limit, color='r')
plt.axvline(upper_limit, color='r')

plt.text(lower_limit+0.01, 450, f'bootstrap lower = {lower_limit.round(2)}', color='r', rotation = 90)
plt.text(upper_limit+0.01, 450, f'bootstrap upper = {upper_limit.round(2)}', color='r', rotation = 90)

plt.axvline(lower_limit_eq, color='g')
plt.axvline(upper_limit_eq, color='g')

plt.text(lower_limit_eq-0.03, 450, f'equation lower = {lower_limit_eq.round(2)}', color='g', rotation = 90)
plt.text(upper_limit_eq-0.03, 450, f'equation upper = {upper_limit_eq.round(2)}', color='g', rotation = 90);
```

### A/B testing

```python
# Perfom a hypothesis testing that checks if there is a significant difference 
# between two of the temperature columns: temp and atemp. 
# Use the scipy.stats and t-test statistics.

# a fast comparison from the entire df

sps.ttest_ind(df['temp'],df['atemp'], nan_policy='omit', equal_var=False)

# highly significant!
```
